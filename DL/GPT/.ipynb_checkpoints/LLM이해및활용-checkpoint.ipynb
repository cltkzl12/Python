# Large Language Model (LLM)의 큰그림 이해 하기
* LLM, 생성형 AI (Generative AI), GPT (Generative Pre-Trained Transformer), Chat GPT라는 여러 단어들이 혼돈을 준다 가볍게 정리를 해보자
* Google Cloud Tech, Introduction to Large Language Models 세션에서 LLM과 Generative AI는 다르지만 교차한다고 설명

<br>
<br>


## LLM (Large Language Model)  

	pretrained(사전 훈련)
	강아지에게 행동을 학습한다고 가정을 했을때 앉아,물어,일어서를 가르치는 과정을 거칩니다. 이후 이 3가지의 행동은 다른 동물들에게도 똑같이
	학습을 시킬수 있기때문에 동일하게 학습을 진행을 합니다. (큰데이터셋으로 사전 훈련)

	fine tuned(미세 조정)
	경찰견, 탐지견과 같은 특정행동을 취해야만하는 경우를 학습하고 싶을떄를 미세 조정이라고 부름 (작은데이터 셋으로 파인 튜닝 )

	이두가지를 거친후에 대규모 범용 언어 모델이 탄생하게 됨

<br>
<br>


## 생성형 AI (Generative AI)


<br>
	간단히 말하면 생성형 AI란 사람이 AI에게 특정 입력(*프롬프트)을 통해 어떤 것을 만들어 달라고 요구하면 그 요구에 맞춰서 결과를 만들어주는 		인공지능을 말합니다. <br>
	텍스트, 오디오, 이미지, 합성 데이터 등 새로운 콘텐츠를 생산할 수 있는 일종의 인공지능 이다.

# OpenAI
## GPT
### GPT 정보
- 대규모 범용 언어 & 자기 회귀 언어 모델
- 고도로 발달한 ‘다음 낱말 맞추기 기계’ and ‘다음 문장 맞추기 기계’

### Chat GPT 정보
* 대화에 특화된 모델
* ‘질문-답’ 대본을 학습한 ‘다음 문장 맞추기 기계’

#### 공통점 
* 두 모델 모두 텍스트 입력에 대해 인간과 같은 응답을 생성할 수 있으며 둘 다 챗봇 및 대화형 AI시스템과 같은 작업이 가능

#### 차이점
* ChatGPT는 대화형 작업을 위해 특별히 설계된 반편 GPT-3는 광범위한 언어 관련 작업에 사용할 수 있는 범용적인(번역 및 코드작성 등) 모델
* ChatGPT는 GPT-3에 비해 더 작은 양의 데이터로 훈련되어 다양하고 미묘한 응답을 생성하는 능력에 차이가 있음.
* GPT-3는 ChatGPT보다 훨씬 더크고 강력하며, ChatGPT의 매개변수가 15억개에 불과한 데 비해 1,750억개의 매개변수가 존재

# 나만의 GPT 모델 만들기 (모델 Fine Tuning)
* GPT와 같은 언어모델을 적용하는 과제가 눈앞에 있는데, 사용자는 어떤 데이터를 준비하면 활용이 가능한지 적용 튜토리얼을 보자
##### Fine Tuning을 하는 이유
> - 서비스에 적용을 해야하는데 모델이 우리 서비스에 대한 정보가 없다면 질문에 답을 할 수 없다.
> - 그래서 우리는 Fine-Tuning과정이 필요하고, 답을 알려주거나 원하는 format으로 답이 나오도록 유도하게끔 하는 효과가 있음.

## Fine Tuning Process
> * 문제 정의
> * 데이터 준비
> * 모델 생성
> * 모델 활용

### 문제 정의
* GPT-3.5 기준  
> 질문 : 웃으면서 말해줘. 너의 이름은?    
> 대답 : 내이름은 홍길동 입니다.  

##### 웃으면서 말해달라고 질문 했지만 이름만 말하는 문제 발견 -> Fine Tuning 진행  


### 데이터 준비

* JSONL 포맷
* 아래와 같이 "prompt"(질문), "completion"(대답) 형태로 데이터 준비

### 모델 생성

# GPT4ALL

## 모델 설명

메타에서 파라미터 크기별로 "7B, 13B, 33B, 65B"의 파라미터를 가지는 4가지 버전의 모델을 공개했다.  
우리가 사용할 GPT4ALL모델은 LLaMA의 7B모델을 Fine Tuning한 모델이다.  

##### NOMIC의 Fine-Tuning을 위한 prompt-response(질답쌍) 제작 방법
> 1. 질문 수집 (LAION OIG(이미지 텍스트쌍), StackoverFlow Questions, Big science/P3)
> 2. GPT-3.5-Turbo를 이용한 답변 제작
> 3. 질문-답변쌍 cleaning 작업

##### 모델 학습
> 1. LLaMB 7B Fine-Tuning  
> 2. LoRA(Low-Rank Adaptation) 적용   

>> * 거대한 모델을 fine-tuning하면 그 엄청난 parameter들을 다 재학습시켜야 함 → 이는 계산량도 많고 시간도 꽤 걸리는 부담스러운 작업이다.
>> * 원래 parameter는 freeze시키고 layer마다 학습가능한 matrices를 추가하는 방식을 도입하였다(Low-Rank Adaptation).
<br>
3. 질답쌍 학습 시작(4 epoch, 4일간 학습)

<br>


* 주의사항
> - 비상업적 라이센스를 갖는 LLaMA기반으로 만들어 졌다.
> - GPT4All 모델 연구들은 finetuning용 prompt-generation 쌍을 생성하기 위해 OpenAI의 GPT-3.5-Turbo를 사용하였다. <br>
OpenAI는 자신들의 제품을 이용하여 OpenAI와 상업적으로 경쟁하는 모델을 개발하는 것을 금지하고 있다.
